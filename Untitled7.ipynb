```python
import pandas as pd
import numpy as np
import math
import tensorflow as tf
import keras
from keras.models import Model
from keras.layers import Input
from keras.layers import Dense
import tensorflow_probability as tfp
from keras.utils import plot_model
tfd = tfp.distributions
from tensorflow.keras.layers import Lambda
```

```python
# read data and check types
dat_train = pd.read_csv("/content/health_train_data.csv")
dat_train.head()
```

```python
# read data and check types
dat_test = pd.read_csv("/content/health_test_data.csv")
dat_test.head()
```

```python
convert_dict = {'hypertension': float,
          'heart_disease': float,
          'ever_married': float,
          'Residence_type': float,
          'stroke': float
                }
dat_train = dat_train.astype(convert_dict)
dat_test = dat_test.astype(convert_dict)
print(dat_train.dtypes)
print(dat_test.dtypes)
```

```python
# create dummy variables
dat_train['Female'] = pd.get_dummies(dat_train.gender)['Female']
print(dat_train.dtypes)
```

```python
xnames = ['age','hypertension','heart_disease','ever_married','Residence_type','avg_glucose_level','bmi','stroke']
ynames = ['health_bills']
xtrain = dat_train[xnames]
ytrain = dat_train[ynames]
```

```python
# define the NN structure
# mean parameter
in_num_layer = Input(shape = (xtrain.shape[1],))
dense1 = Dense(3, activation = 'tanh')(in_num_layer)
mu = Dense(1, activation = 'exponential')(dense1)

# dispersion parameter
in_fix_layer = Input(shape = (1, ))
phi = Dense(1, activation = 'exponential', use_bias=False)(in_fix_layer)

# input
in_layers = [in_num_layer, in_fix_layer]

# gamma pdf
#py = tfp.layers.DistributionLambda(lambda t: tfd.Gamma(t[0]/t[1], 1/t[1]))((mu, phi))
py = tfp.layers.DistributionLambda(lambda t: tfd.Gamma(1/t[1], 1/t[0]*t[1]))((mu, phi))
```

```python
model = Model(inputs = in_layers, outputs = py)
model.summary()
```

```python
plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=False, dpi=96)
```

```python
# define loss function
#negloglik = lambda y_true, y_hat: -y_hat.log_prob(tf.cast(y_true, tf.float32))  
negloglik = lambda y_true, y_hat: -y_hat.log_prob(y_true)  
model.compile(optimizer = 'adam', loss = negloglik)
```

```python
xnames = ['age','hypertension','heart_disease','ever_married','Residence_type','avg_glucose_level','bmi','stroke']
ynames = ['health_bills']
xtest = dat_test[xnames]
ytest = dat_test[ynames]
```

```python
#one_vector = np.array([1] * xtrain.shape[0])  
one_vector = np.ones(xtest.shape[0])
model.fit([xtest, one_vector], ytest, epochs=250, shuffle=False, batch_size=256, verbose=2)
```

```python
# prediction
x_new = [xtest, one_vector]
yhat = model.predict(x_new)
print(yhat.mean())
```

```python
# calculate test MSE
from sklearn.metrics import mean_absolute_error
mean_absolute_error(yhat,ytest)
```
