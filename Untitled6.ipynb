```python
# build unigram
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
# normalization
import unicodedata
# stopwords
from nltk.corpus import stopwords
# punctuation
import string
# clean using regex expression
import re
#stemming
from nltk.stem.snowball import SnowballStemmer
# lemmatize
from nltk.stem import WordNetLemmatizer
```

```python
dat = pd.read_csv("WorkersComp.csv")  
dat.head()
```

```python
nltk.download('stopwords')
sw = set(stopwords.words("english"))
```

```python
nltk.download('wordnet')
stemmer = SnowballStemmer("english")
lemmatizer = WordNetLemmatizer()
Stem = []
Lemm = []
for Claim in dat.ClaimDescription:
    text = Claim.lower() # lowerlize
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')# not neccessary
    text = text.split()
    tokens_stemmed=[SnowballStemmer("english").stem(i) for i in text if (i not in string.punctuation)&(i not in sw)]
    Stem.append(tokens_stemmed)
    tokens_lemmatized=[lemmatizer.lemmatize(i) for i in tokens_stemmed if (i not in string.punctuation)&(i not in sw)]
    Lemm.append(tokens_lemmatized)
```

word embeedings from 'glove-twitter-25'

```python
import gensim
import gensim.downloader as api
model = api.load("glove-twitter-25")
```

```python
words = ['tea', 'phone', 'coffee', 'moon']
for word in words:
    if word in model:
        print('Embedding vector for {}:'.format(word))
        print(model[word])
        print('\n')
    else:
        print('{} is not in the vocabulary.\n'.format(word))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectors = []
for word in words:
    if word in model:
        vectors.append(model[word])
    else:
        print('{} is not in the vocabulary.\n'.format(word))

# Calculate the similarity matrix
similarity_matrix = cosine_similarity(vectors)

# Find the indices of the pair with the highest similarity
max_similarity_index = np.unravel_index(similarity_matrix.argmax(), similarity_matrix.shape)

# Get the words corresponding to the indices
word1 = words[max_similarity_index[0]]
word2 = words[max_similarity_index[1]]

# Print the similarity matrix and the pair with the highest similarity
print('Similarity matrix:')
print(similarity_matrix)
print('\n')
print('Pair with highest similarity: {} and {}, similarity score = {}'.format(word1, word2, similarity_matrix[max_similarity_index]))
```
